```python
from pyspark.sql.functions import col, substring

staging_path = "Files/stage/"
bronze_path = 'Files/bronze/'
tb_servicios_path = 'transacciones_servicios/'

tr_servicios = "transacciones_servicios_"

#variable que tendra el path del archivo a cargar
path_archivo_tr_servicios = staging_path + tr_servicios + ".csv"

#path para transacciones terceros en bronze
tr_servicios_bronze_path = bronze_path + tb_servicios_path

print(path_archivo_tr_servicios)
print(tr_servicios_bronze_path)
```

```python
df_tr_servicios = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("delimiter", ",") \
    .option("quote", '"') \
    .option("escape", '\\') \
    .option("multiLine", "true") \
    .csv(path_archivo_tr_servicios)
```

```python
df_tr_servicios.printSchema()
```

```python
df_tr_servicios.write\
                .format('delta')\
                .mode('append')\
                .save(tr_servicios_bronze_path)
```
