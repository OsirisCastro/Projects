```python
#Configurations
from pyspark.sql.functions import*

staging_path = "Files/"
dir_tr_servicios = "transacciones_servicios_"

#variable que tendra el path del archivo a cargar
path_archivo = staging_path + dir_tr_servicios + ".csv"
```

```python
df = spark.read.format("binaryFile").load("Files/stage/")
df.select("path").show(truncate=False)
```

```python
df_data = spark.read \
    .option("header", "true") \
    .option("delimiter", ",") \
    .option("quote", '"') \
    .option("escape", '"') \
    .option("multiLine", "true") \
    .option("inferSchema", "true") \
    .csv(path_archivo)
```

```python
df_data.printSchema()
```
